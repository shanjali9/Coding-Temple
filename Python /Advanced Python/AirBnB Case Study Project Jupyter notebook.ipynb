```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('airbnb.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

air_bnb['neighbourhood_group'].describe()

# There are five available neighborhood groups & Manhattan is the most common with a frequency of 21,661 
```

```python
# Are private rooms the most popular in manhattan?

air_bnb.groupby(['neighbourhood_group', 'room_type']).count().sort_values('neighbourhood_group', ascending = False)

#No, they are not. Entire home/apt is.
```

```python
# Which hosts are the busiest and based on their reviews?

air_bnb.groupby(['host_name', 'reviews_per_month']).count().sort_values('reviews_per_month', ascending = False)

# Row NYC, Louann, and Naicia if counted by reviews per month. If counted by total number, Dona and Jj.
```

```python
#Which neighorhood group has the highest average price?

air_bnb.groupby('neighbourhood_group').mean()[['price']]

#Manhattan 
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.groupby('neighbourhood_group').describe()[['price']]

#Manhattan has the highest total price.
```

```python
#Which top 5 hosts have the highest total price?

air_bnb.groupby(['host_name', 'price']).head().sort_values('price', ascending = False)

# Jelena, Erin, Kathrine, Matt, and Olson.
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

grouped = air_bnb[air_bnb['number_of_reviews'] >= 100].groupby(['host_name', 'availability_365']).head().sort_values('availability_365', ascending = True)
grouped[grouped['availability_365'] == 0]

#162 users as below 
```

```python
# What host has the highest total of prices and where are they located?

grouped = (air_bnb[air_bnb['host_name'] == 'Sonder (NYC)'])
grouped_copy = grouped.drop_duplicates('neighbourhood').reset_index(drop=True)
grouped_copy['neighbourhood']

# Sonder (NYC). It is in the neighbourhood group Manhattan and operates in the Financial District.
```

```python
# When did Danielle from Queens last receive a review?

air_bnb[air_bnb['host_id'] == 26432133]

# Her last review was on July 7th, 2019.
```
